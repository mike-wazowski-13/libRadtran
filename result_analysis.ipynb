```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results30 = pd.read_csv(r"libRad30_variability.csv")
results45 = pd.read_csv(r"libRad45_variability.csv")
results60 = pd.read_csv(r"libRad60_variability.csv")
```

```python
magnitude = "angstrom"
plt.figure(figsize=(12,6))
df_30 = results30[results30["variable"]==magnitude]
df_45 = results30[results45["variable"]==magnitude]
df_60 = results30[results60["variable"]==magnitude]

plt.plot(df_30["value"], df_30["Glob_dif"], label="Global")
plt.plot(df_30["value"], df_30["Dir_dif"], label="Direct")
plt.plot(df_30["value"], df_30["Difu_dif"], label="Diffuse")
plt.legend()
plt.xlabel(f"{magnitude}")
plt.ylabel("% of change")
plt.title("Irradiance variation from the standard case")
```

```python
dfs = {
    "30": df_30,
    "45": df_45,
    "60": df_60
}
plt.figure(figsize=(12,6))
for angle, df in dfs.items():
    plt.plot(df["value"], df["Glob_dif"], label=f"Global {angle}")
    plt.plot(df["value"], df["Dir_dif"],  label=f"Direct {angle}")
    plt.plot(df["value"], df["Difu_dif"], label=f"Diffuse {angle}")

plt.xlabel("Value")
plt.ylabel("Difference")
plt.legend()
plt.show()
```
