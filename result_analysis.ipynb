```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({
    "axes.labelsize": 14,
    "xtick.labelsize": 12,
    "ytick.labelsize": 12,
    "legend.fontsize": 12,
    "axes.titlesize": 16
})
```

```python
results30 = pd.read_csv(r"libRad30_variability.csv")
results45 = pd.read_csv(r"libRad45_variability.csv")
results60 = pd.read_csv(r"libRad60_variability.csv")
```

```python
intg_results30 = pd.read_csv(r"libRad30_intg_diff.csv")
intg_results45 = pd.read_csv(r"libRad45_intg_diff.csv")
intg_results60 = pd.read_csv(r"libRad60_intg_diff.csv")
```

```python
magnitude = "DOY"

df_30 = results30[results30["variable"]==magnitude]
df_45 = results45[results45["variable"]==magnitude]
df_60 = results60[results60["variable"]==magnitude]
```

```python
plt.figure(figsize=(12,6))
plt.plot(df_30["value"], df_30["Glob_dif"], label="Global")
plt.plot(df_30["value"], df_30["Dir_dif"], label="Direct")
plt.plot(df_30["value"], df_30["Difu_dif"], label="Diffuse")
plt.legend()
plt.xlabel(f"{magnitude}")
plt.ylabel("% of change")
plt.title("Irradiance variation from the standard case")
```

```python
dfs = { "30": df_30, "45": df_45, "60": df_60 }
colors = {
    "global":  "#1f77b4",  # blue-ish
    "direct":  "#ff7f0e",  # orange-ish
    "diffuse": "#2ca02c"   # green-ish
}

linestyles = {
    "30": "-",
    "45": "--",
    "60": ":"
}
```

```python
plt.figure(figsize=(12, 6))
for angle, df in dfs.items():
    ls = linestyles[angle]

    plt.plot(df["value"], df["Glob_dif"],
             color=colors["global"], linestyle=ls,
             label=f"Global {angle}")

    plt.plot(df["value"], df["Dir_dif"],
             color=colors["direct"], linestyle=ls,
             label=f"Direct {angle}")

    plt.plot(df["value"], df["Difu_dif"],
             color=colors["diffuse"], linestyle=ls,
             label=f"Diffuse {angle}")

plt.xlabel(magnitude)
plt.ylabel("% of change")
plt.title("Irradiance variation from the standard case")
plt.legend()
```

```python
df_30 = intg_results30[intg_results30["variable"]==magnitude]
df_45 = intg_results45[intg_results45["variable"]==magnitude]
df_60 = intg_results60[intg_results60["variable"]==magnitude]

plt.figure(figsize=(12,6))
plt.plot(df_30["value"], df_30["Glob_dif"], label="Global")
plt.plot(df_30["value"], df_30["Dir_dif"], label="Direct")
plt.plot(df_30["value"], df_30["Difu_dif"], label="Diffuse")
plt.legend()
plt.xlabel(f"{magnitude}")
plt.ylabel("Irradiance [W/m2]")
plt.title("Integrated Irradiance difference from the standard case")
```
